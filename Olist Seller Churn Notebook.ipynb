## **1. Domain Knowledge**

### **1.1. Background Knowledge**

Olist is a Brazilian e-commerce enablement platform founded in 2015 that specializes in connecting small and medium sized retailers with major online marketplaces. Instead of operating as a traditional marketplace where buyers purchase directly from a single platform.

Olist acts as an intermediary that empowers merchants to sell their products across multiple sales channels. Sellers can list products, manage inventory, process orders, track logistics, and receive payments all from one centralized interface. Olist also collaborates with logistics partners to facilitate delivery and offers financial services that support merchant operations.

Olist operates two main solutions, Olist Store and Olist VNDA. Olist Store is a marketplace integrator that allows sellers to list products once and have them automatically published across multiple major marketplaces under the “Olist Store” branding, and Olist VNDA, an e-commerce platform that enables brands to build and manage their own direct to consumer online store.

And the dataset that will be used in this project is an Olist Store Dataset from [kaggle.com](https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce)

### **1.2. How it Works**

The service flow of Olist can be summarized as follows :
- The seller registers on Olist Store.

- Olist provides a catalog and dashboard to manage the shop, products, and orders.

- The sellers products are automatically displayed across multiple marketplaces (Mercado Livre, Americanas, Submarino, Walmart Brazil, etc.)

- Buyers place their orders on the marketplace, where the product may actually be operated and fulfilled through Olist.

- The seller ships the products to the logistics center.

- The seller sends the items to the assigned carrier/logistics aggregator according to Olist instructions.

- Olist handles the delivery to the final customer, integrating with various logistics providers.

- Olist settles the payment to the seller after the order is completed and no return/refund issues remain.

The data flow within the Olist dataset is as follows :
- Product information -> `product_category_name`, `product size/photos/weight`

- Customer makes the purchase -> `customer_id`, `customer location`

- Product items added to the order -> `order_item_id`, `product_id`, `seller_id`, `price`, `freight_value`

- Payment is processed -> `payment_sequential`, `payment_type`, `payment_installments`, `payment_value`

- Order creation -> `order_id`, `order_purchase_timestamp`, `order_status`

- Shipping is processed -> `order_approved_at`, `order_delivered_carrier_date`, `order_estimated_delivery_date`

- Item delivered to the customer -> `order_delivered_customer_date`

- Customer leaves a review (optional) -> `review_score`, `review_comment_title`, `review_comment_message`

Olist revenues :
- Subscription fees (Monthly subscription) [*Olist Plan Pricing*](https://olist.com/planos/)

**Checking for timeline inconsistencies**

We will validate each order event to ensure that the sequence of dates follows a logical process.

```python
time_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'review_creation_date',
    'review_answer_timestamp',
    'order_estimated_delivery_date'
]

for i, col in enumerate(time_cols):
    for next_col in time_cols[i+1:]:
        mask = df[df[col] > df[next_col]].index
        count = len(mask)
        print(f'{col} > {next_col} = {count} rows')
```

- `order_purchase_timestamp > order_delivered_carrier_date`
- `order_purchase_timestamp > review_creation_date`
- `order_purchase_timestamp > review_answer_timestamp`
- `order_approved_at > order_delivered_carrier_date`
- `order_approved_at > order_delivered_customer_date`
- `order_approved_at > review_creation_date`
- `order_approved_at > review_answer_timestamp`
- `order_delivered_carrier_date > order_delivered_customer_date`
- `order_delivered_carrier_date > review_creation_date`
- `order_delivered_carrier_date > review_answer_timestamp`

```python
correction_log = {}

def log_fix(issue, count):
    if count > 0:
        correction_log[issue] = count

# order_purchase_timestamp > order_delivered_carrier_date
mask = df['order_purchase_timestamp'] > df['order_delivered_carrier_date']
log_fix("order_purchase_timestamp > order_delivered_carrier_date", mask.sum())
df.loc[mask, 'order_delivered_carrier_date'] = df.loc[mask, 'shipping_limit_date']

# order_purchase_timestamp > review_creation_date
mask = df['order_purchase_timestamp'] > df['review_creation_date']
log_fix("order_purchase_timestamp > review_creation_date", mask.sum())
df.loc[mask, 'review_creation_date'] = df.loc[mask, 'order_delivered_customer_date']

# order_purchase_timestamp > review_answer_timestamp
mask = df['order_purchase_timestamp'] > df['review_answer_timestamp']
log_fix("order_purchase_timestamp > review_answer_timestamp", mask.sum())
df.loc[mask, 'review_answer_timestamp'] = df.loc[mask, 'review_creation_date']

# order_approved_at > order_delivered_carrier_date
mask = df['order_approved_at'] > df['order_delivered_carrier_date']
log_fix("order_approved_at > order_delivered_carrier_date", mask.sum())
df.loc[mask, 'order_delivered_carrier_date'] = df.loc[mask, 'order_approved_at'] + pd.Timedelta(days=1)

# order_approved_at > order_delivered_customer_date
mask = df['order_approved_at'] > df['order_delivered_customer_date']
log_fix("order_approved_at > order_delivered_customer_date", mask.sum())
df.loc[mask, 'order_approved_at'] = df.loc[mask, 'order_purchase_timestamp'] + pd.Timedelta(days=1)

# order_approved_at > review_creation_date
mask = df['order_approved_at'] > df['review_creation_date']
log_fix("order_approved_at > review_creation_date", mask.sum())
df.loc[mask, 'review_creation_date'] = df.loc[mask, 'order_delivered_customer_date'] + pd.Timedelta(days=1)

# order_approved_at > review_answer_timestamp
mask = df['order_approved_at'] > df['review_answer_timestamp']
log_fix("order_approved_at > review_answer_timestamp", mask.sum())
df.loc[mask, 'review_answer_timestamp'] = df.loc[mask, 'review_creation_date']

# order_approved_at > order_estimated_delivery_date
mask = df['order_approved_at'] > df['order_estimated_delivery_date']
log_fix("order_approved_at > order_estimated_delivery_date", mask.sum())
df.loc[mask, 'order_estimated_delivery_date'] = df.loc[mask, 'order_delivered_customer_date']

# order_delivered_carrier_date > order_delivered_customer_date
mask = df['order_delivered_carrier_date'] > df['order_delivered_customer_date']
log_fix("order_delivered_carrier_date > order_delivered_customer_date", mask.sum())
df.loc[mask, 'order_delivered_customer_date'] = df.loc[mask, 'order_delivered_carrier_date'] + pd.Timedelta(days=7)

# order_delivered_carrier_date > review_creation_date
mask = df['order_delivered_carrier_date'] > df['review_creation_date']
log_fix("order_delivered_carrier_date > review_creation_date", mask.sum())
df.loc[mask, 'review_creation_date'] = df.loc[mask, 'order_delivered_customer_date']

# order_delivered_carrier_date > review_answer_timestamp
mask = df['order_delivered_carrier_date'] > df['review_answer_timestamp']
log_fix("order_delivered_carrier_date > review_answer_timestamp", mask.sum())
df.loc[mask, 'review_answer_timestamp'] = df.loc[mask, 'review_creation_date']

# order_delivered_customer_date > review_creation_date
mask = df['order_delivered_customer_date'] > df['review_creation_date']
log_fix("order_delivered_customer_date > review_creation_date", mask.sum())
df.loc[mask, 'review_creation_date'] = df.loc[mask, 'order_delivered_customer_date'] + pd.Timedelta(days=1)

# order_delivered_customer_date > review_answer_timestamp
mask = df['order_delivered_customer_date'] > df['review_answer_timestamp']
log_fix("order_delivered_customer_date > review_answer_timestamp", mask.sum())
df.loc[mask, 'review_answer_timestamp'] = df.loc[mask, 'review_creation_date'] + pd.Timedelta(days=1)

# order_delivered_carrier_date > order_estimated_delivery_date
mask = df['order_delivered_carrier_date'] > df['order_estimated_delivery_date']
log_fix("order_delivered_carrier_date > order_estimated_delivery_date", mask.sum())
df.loc[mask, 'order_estimated_delivery_date'] = df.loc[mask, 'order_delivered_carrier_date'] + pd.Timedelta(days=7)
```

### **3.5. Create New Data**

We will create a new dataset using the variables that we consider most relevant for both the analysis and the development of the model. This refined dataset will help ensure clearer insights, reduce noise from unnecessary features, and improve the overall performance and accuracy of the machine learning model.

Defining Churn based on sellers acitivity on each quarter

```python
df['order_quarter'] = df['order_purchase_timestamp'].dt.to_period('Q').astype(str)
```

Creating `total_orders` per quarter

```python
churn_df = df.groupby(['seller_id', 'order_quarter']).agg(total_orders=('order_id', 'nunique')).reset_index()
churn_df
```

Creating `cumulative_orders` from its quarter

```python
churn_df.sort_values(['seller_id', 'order_quarter'])
churn_df['cumulative_orders'] = churn_df.groupby('seller_id')['total_orders'].cumsum()
churn_df
```

Creating `next_quarter` for defining churn label

```python
churn_df['next_quarter'] = churn_df.groupby('seller_id')['order_quarter'].shift(-1)
churn_df['churn'] = churn_df['next_quarter'].isna().astype(int)
churn_df
```

Creating `monetary` columns based on total sales on each seller quarter

```python
quarter_monetary = df.groupby(['seller_id', 'order_quarter']).agg(monetary=('price', 'sum')).reset_index()

churn_df = churn_df.merge(quarter_monetary, on=['seller_id', 'order_quarter'], how='left')
churn_df
```

Creating `cumulative_monetary`

```python
churn_df['cumulative_monetary'] = churn_df.groupby('seller_id')['monetary'].cumsum()
churn_df
```

Creating `recency_days` how many days from last transaction on each quarter

```python
quarter_last_tx = df.groupby(['seller_id', 'order_quarter'])['order_purchase_timestamp'].max().reset_index()
quarter_last_tx.rename(columns={'order_purchase_timestamp': 'last_tx_date'}, inplace=True)
quarter_last_tx['order_quarter'] = quarter_last_tx['order_quarter'].astype('period[Q]')
quarter_last_tx['quarter_end_date'] = quarter_last_tx['order_quarter'].dt.to_timestamp(how='end')
quarter_last_tx['recency_days'] = (quarter_last_tx['quarter_end_date'] - quarter_last_tx['last_tx_date']).dt.days

churn_df['order_quarter'] = churn_df['order_quarter'].astype('period[Q]')
churn_df = churn_df.merge(
    quarter_last_tx.loc[:, ['seller_id', 'order_quarter', 'recency_days']],
    on=['seller_id', 'order_quarter'], how='left')
churn_df
```

Creating `active_months_count` = total month of seller was active

```python
df['active_months'] = df['order_purchase_timestamp'].dt.to_period('M')

active_months = df.groupby(['seller_id', 'order_quarter'])['active_months'].nunique().reset_index()
active_months.rename(columns={'active_months': 'active_months_count'}, inplace=True)
active_months['order_quarter'] = active_months['order_quarter'].astype('period[Q]')

churn_df = churn_df.merge(active_months, on=['seller_id', 'order_quarter'], how='left')
churn_df
```

Creating `seller_active_days_count` total days of seller was active

```python
df['order_date'] = df['order_purchase_timestamp'].dt.date

active_days = df.groupby(['seller_id', 'order_quarter'])['order_date'].nunique().reset_index(name='active_days')
active_days.rename(columns={'active_days': 'active_days_count'}, inplace=True)
active_days['order_quarter'] = active_days['order_quarter'].astype('period[Q]')

churn_df = churn_df.merge(active_days,on=['seller_id', 'order_quarter'],how='left')
churn_df
```

Creating `tenure_months` the length of time the seller has been active since the first transaction

```python
first_tx = df.groupby('seller_id')['order_purchase_timestamp'].min().reset_index()
first_tx.rename(columns={'order_purchase_timestamp': 'first_tx_date'}, inplace=True)

churn_df = churn_df.merge(first_tx, on='seller_id', how='left')
churn_df['quarter_end_date'] = churn_df['order_quarter'].dt.end_time
churn_df['tenure_months'] = (churn_df['quarter_end_date'] - churn_df['first_tx_date']).dt.days / 30
churn_df['tenure_months'] = churn_df['tenure_months'].round(1)
churn_df.drop(columns=['first_tx_date', 'quarter_end_date'], inplace=True)
churn_df
```

Getting seller location data from previous dataframe

```python
seller_location = df[['seller_id', 'seller_city', 'seller_state']].drop_duplicates()

churn_df = churn_df.merge(seller_location, on='seller_id', how='left')
churn_df
```

## **5. Modeling**

### **5.1. Introduction**

Because Olist is a platform primarily oriented toward sellers, and its main source of revenue is subscription fees, the sellers become the core customers of Olist. Therefore, it is crucial for Olist to minimize seller churn as much as possible. With a minimized churn rate, we can prevent revenue losses also identifying key strategies to increase profitability.

Therefore, the model we aim to build is expected to help Olist determine the appropriate interventions for each seller. By identifying sellers who are at high risk of churn, also Olist will be able to apply proactive retention strategies.

### **5.2. Data Understanding**

The data used to build the machine learning model contains the following fields:

| **order_quarter**                | Quarter in which the seller was active (period format Q, e.g., 2017Q3)                                                                  |

### **5.3. Metrics Evaluation**

|                        | **Predicted Active (0)** | **Predicted Churn (1)** |
| ---------------------- | ------------------------ | ----------------------- |
| **Actual Active (0)**  |   True Negative (TN)     |  False Positive (FP)    |
| **Actual Churn  (1)**  |   False Negative (FN)    |  True Positive (TP)     |

TP (True Positive) = The model correctly predicts sellers who churn.

FP (False Positive) = The model predicts that a seller will churn, but the seller actually remains active.

TN (True Negative) = The model correctly predicts sellers who remain active.

FN (False Negative) = The model fails to predict sellers who churn.

#### **5.3.1. Segmentation**

Before planning the best intervention, we will define which intervention should we use for each seller by creating seller segmentation

```python
churn_df = pd.read_csv('Churn Data.csv')
ml_df = churn_df.copy()
ml_df
```

We will create sellers segment using the LRFM framework, and the LRFM values are based on tenure_months, recency_days, cumulative_orders, and cumulative_monetary

```python
ml_df2 = ml_df.copy()

ml_df2['L'] = pd.qcut(ml_df2['tenure_months'], 4, labels=[1,2,3,4]).astype(int)
ml_df2['R'] = pd.qcut(ml_df2['recency_days'], 4, labels=[4,3,2,1]).astype(int)
ml_df2['F'] = pd.qcut(ml_df2['cumulative_orders'], 4, labels=[1,2,3,4]).astype(int)
ml_df2['M'] = pd.qcut(ml_df2['cumulative_monetary'], 4, labels=[1,2,3,4]).astype(int)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

columns = ['L', 'R', 'F', 'M']
positions = [(0,0), (0,1), (1,0), (1,1)]

for col, pos in zip(columns, positions):
    axes[pos].boxplot(ml_df2[col])
    axes[pos].set_title(f'Boxplot {col}')

plt.tight_layout()
plt.show()
```

```python
lrfm = ['L', 'R', 'F', 'M']

for i in lrfm:
    print(f'Skewness {i}:', ml_df2[i].skew())
```

Because the data distribution is skewed, we will use K-Medoids

```python
X = ml_df2[['L', 'R', 'F', 'M']].copy()
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

Since Olist’s pricing segmentation also consists of four tiers, we will divide the seller segments into four groups as well

```python
from pyclustering.cluster.kmedoids import kmedoids

k_cluster = 4

np.random.seed(0)
initial_medoids = np.random.choice(len(X_scaled), k_cluster, replace=False).tolist()

kmedoids_final = kmedoids(
    X_scaled.tolist(),
    initial_medoids,
    data_type='points'
)
kmedoids_final.process()

clusters = kmedoids_final.get_clusters()
medoids = kmedoids_final.get_medoids()
```

```python
labels = np.zeros(len(X_scaled), dtype=int)
for cluster_id, cluster_points in enumerate(clusters):
    for point_idx in cluster_points:
        labels[point_idx] = cluster_id

ml_df2['cluster'] = labels
```

```python
ml_df2.groupby('cluster')[['L','R','F','M']].median()
```

We classified the sellers and determined which segment each cluster belongs to based on their LRFM values

| Cluster | Characteristic Pattern                                              | Segment Name                                                                                |
|---------|---------------------------------------------------------------------|---------------------------------------------------------------------------------------------|
| 0       | High L, Very Low R (inactive recently), Low–Medium F, Low–Medium M | **Tier 2 : Inactive/Occasional Customers** : long time customers but currently inactive or low use  |
| 1       | Low L, High R (recent activity), Very Low F, Very Low M            | **Tier 1 : New At Risk Customers** → newly acquired customers with very low activity                |
| 2       | Very High L, Medium R, Very High F, Very High M                    | **Tier 4 : VIP Loyal Customers** → long-term, highly active, and highest-spending customers         |
| 3       | Medium L, High R, High F, High M                                   | **Tier 3 : Loyal Regular Customers** -> consistent, active customers with strong purchase behavior    |

```python
segment_map = {
    0: 'Tier 2',
    1: 'Tier 1',
    2: 'Tier 4',
    3: 'Tier 3'
}

ml_df2['segment'] = ml_df2['cluster'].map(segment_map)
ml_df2[['seller_id', 'order_quarter', 'churn', 'segment']]
```

```python
def create_baseline_cm(data, segments, prediction_value=0):
    cms = {}
    for seg in segments:
        actual = data.query(f'segment == "{seg}"')['churn']
        if prediction_value == 0:
            predicted = np.zeros(actual.shape[0])
        else:
            predicted = np.ones(actual.shape[0])
        cms[seg] = confusion_matrix(actual, predicted)
    return cms

segments_list = ['Tier 1', 'Tier 2', 'Tier 3', 'Tier 4']

all_cm0 = create_baseline_cm(test_data, segments_list, prediction_value=0)
all_cm1 = create_baseline_cm(test_data, segments_list, prediction_value=1)

cm0_t1 = all_cm0['Tier 1']
cm0_t2 = all_cm0['Tier 2']
cm0_t3 = all_cm0['Tier 3']
cm0_t4 = all_cm0['Tier 4']

cm1_t1 = all_cm1['Tier 1']
cm1_t2 = all_cm1['Tier 2']
cm1_t3 = all_cm1['Tier 3']
cm1_t4 = all_cm1['Tier 4']
```

```python
def analyze_cost(cm, segment):
    tn, fp, fn, tp = cm.reshape(-1)

    tier_setting = {
        "Tier 1": (59, 10),
        "Tier 2": (159, 30),
        "Tier 3": (349, 40),
        "Tier 4": (849, 100),
    }
    tier_revenue, tier_cost = tier_setting[segment]

    fp_cost = -tier_cost * fp
    fn_cost = -tier_revenue * fn

    total_cost = fp_cost + fn_cost

    return fp_cost, fn_cost, total_cost


def analyze_all_cost(all_cm, segments_list):
    results = {}
    totals = {
        'fp_cost': 0,
        'fn_cost': 0,
        'total_cost': 0
    }
    
    for segment in segments_list:
        cm = all_cm[segment]
        fp_c, fn_c, total = analyze_cost(cm, segment)
        
        results[segment] = {
            'fp_cost': fp_c,
            'fn_cost': fn_c,
            'total_cost': total
        }
        
        totals['fp_cost'] += fp_c
        totals['fn_cost'] += fn_c
        totals['total_cost'] += total
    
    results['TOTAL'] = totals
    return results
```

```python
results = []

for model_name, model in models.items():
  split = GroupTimeSeriesSplit(n_splits=3, test_size=1)
  
  model_pipe = Pipeline([
    ('preprocessing', preprocessing),
    ('smote', SMOTE(random_state=0)),
    ('modeling', RFE(estimator=model)),
  ])

  scores = cross_val_score(
    estimator=model_pipe,
    X=X_train,
    y=y_train,
    cv=split,
    groups = X_train.index,
    scoring=f2_score,
  )

  results.append({
    'model': model_name,
    'cv_score': scores.round(2),
    'mean_score': scores.mean().round(4),
    'std_score': scores.std().round(4),
  })

benchmark_results = pd.DataFrame(results).sort_values('mean_score', ascending=False).reset_index(drop=True)
benchmark_results
```

### **5.8. Hyperparameter Tuning**

```python
from hyperopt import fmin, tpe, hp, Trials, space_eval
from hyperopt.pyll.base import scope
```

```python
params_space = {
    'max_depth': scope.int(hp.quniform('max_depth', 3, 12, 1)),
    'n_estimators': scope.int(hp.quniform('n_estimators', 50, 300, 10)),
    'learning_rate': hp.loguniform('learning_rate', np.log(0.01), np.log(0.2)),
    'min_child_weight': scope.int(hp.quniform('min_child_weight', 1, 20, 1)),
    'scale_pos_weight': hp.uniform('scale_pos_weight', 3, 10),
    'max_delta_step': scope.int(hp.choice('max_delta_step', [0, 1, 3, 5])),
    'subsample': hp.choice('subsample', [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]),
    'colsample_bytree': hp.choice('colsample_bytree', [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]),
}
```

```python
def objective(args):
    
    pipe_tuning = Pipeline([
        ('preprocessing', preprocessing),
        ('smote', SMOTE(random_state=0)),
        ('model', RFE(estimator=XGBClassifier(random_state=0, **args)))
    ])

    split = GroupTimeSeriesSplit(n_splits=3, test_size=1)

    scores = cross_val_score(
        estimator=pipe_tuning,
        X=X_train,
        y=y_train,
        cv=split,
        groups=X_train.index,
        scoring=f2_score,
        n_jobs=-1
    )

    return 1 - scores.mean()
```
